```python
X = data.drop(['Result','index'],axis=1)
Y = data["Result"]
```

# Preprocessing

```python
scaler = StandardScaler()
scaler.fit(X)
X = scaler.transform(X)
```

```python
from sklearn.model_selection import train_test_split 
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.3, random_state = 0) 
```

# Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
from sklearn import metrics
logreg = LogisticRegression()
logreg.fit(X_train, Y_train)
Y_pred=logreg.predict(X_test)
```

```python
print(metrics.confusion_matrix(Y_test, Y_pred))
print(metrics.classification_report(Y_test, Y_pred))
print('\nAccuracy Score:')
print(metrics.accuracy_score(Y_test,Y_pred))
```

# Naive Bayes

```python
from sklearn.naive_bayes import GaussianNB
model = GaussianNB()
model.fit(X_train,Y_train)
Y_pred = model.predict(X_test)
```

```python
print(metrics.confusion_matrix(Y_test, Y_pred))
print(metrics.classification_report(Y_test, Y_pred))
print('\nAccuracy Score:')
print(metrics.accuracy_score(Y_test,Y_pred))
```

# Support Vector Machine 

```python
from sklearn.svm import SVC
svc=SVC() #Default hyperparameters
svc.fit(X_train,Y_train)
Y_pred=svc.predict(X_test)
```

```python
print(metrics.confusion_matrix(Y_test, Y_pred))
print(metrics.classification_report(Y_test, Y_pred))
print('\nAccuracy Score:')
print(metrics.accuracy_score(Y_test,Y_pred))
```

## Linear

```python
svc=SVC(kernel='linear')
svc.fit(X_train,Y_train)
Y_pred=svc.predict(X_test)
```

```python
print(metrics.confusion_matrix(Y_test, Y_pred))
print(metrics.classification_report(Y_test, Y_pred))
print('\nAccuracy Score:')
print(metrics.accuracy_score(Y_test,Y_pred))
```

## RBF Kernel

```python
svc=SVC(kernel='rbf')
svc.fit(X_train,Y_train)
Y_pred=svc.predict(X_test)
```

```python
print(metrics.confusion_matrix(Y_test, Y_pred))
print(metrics.classification_report(Y_test, Y_pred))
print('\nAccuracy Score:')
print(metrics.accuracy_score(Y_test,Y_pred))
```

## Polynomial

```python
svc=SVC(kernel='poly')
svc.fit(X_train,Y_train)
Y_pred=svc.predict(X_test)
```

```python
print(metrics.confusion_matrix(Y_test, Y_pred))
print(metrics.classification_report(Y_test, Y_pred))
print('\nAccuracy Score:')
print(metrics.accuracy_score(Y_test,Y_pred))
```

# KNN

```python
from sklearn.neighbors import KNeighborsClassifier
clf = KNeighborsClassifier(n_neighbors = 2)
clf.fit(X_train, Y_train)
Y_pred =clf.predict(X_test)
```

```python
from sklearn.ensemble import RandomForestClassifier
clf = RandomForestClassifier(n_estimators = 30 )
clf.fit(X_train, Y_train)
Y_pred =clf.predict(X_test)
```

```python
Y_pred = model.predict(X_test)
for i in range(len(Y_pred)):
    if Y_pred[i]<0.5:
        Y_pred[i]=-1
    else:
        Y_pred[i]=1        
```

```python
print(metrics.confusion_matrix(Y_test, Y_pred))
print(metrics.classification_report(Y_test, Y_pred))
print('\nAccuracy Score:')
print(metrics.accuracy_score(Y_test,Y_pred))
```

# Representation of Algorithms metrics

```python
accuracy = pd.read_csv("D:\\Academics\\Kavach\\Implementation\\Level 3\\val.csv")
```

```python
accuracy
```

```python
import matplotlib.pyplot as plt

plt.xticks(rotation=90)
bar = plt.bar(accuracy['algorithm'], accuracy['Accuracy'], color='maroon', width=0.4)

plt.xlabel("Algorithm")
plt.ylabel("Accuracy")
plt.title("Algorithm vs Accuracy")

for rect in bar:
    height = rect.get_height()
    plt.annotate('{:.2f}'.format(height),
                 xy=(rect.get_x() + rect.get_width() / 2, height),
                 xytext=(0, 3),
                 textcoords="offset points",
                 ha='center', va='bottom')

plt.subplots_adjust(bottom=0.15)

plt.show()
```

```python
plt.scatter(accuracy['F2Phish'],accuracy['F2NonPhish'],color="blue")
plt.xlabel("Phishing")
plt.ylabel("Non-Phishing")
plt.title("Alogrithms F2 score")
plt.show()
```
