```python
import csv
import requests
from bs4 import BeautifulSoup

# The base URL of the webpage to scrape
base_url = 'https://phishtank.org/phish_search.php?page={}&valid=n&Search=Search'

# Open a CSV file for writing
with open('table_values_invalid.csv', 'w', newline='') as csvfile:
    writer = csv.writer(csvfile)
    writer.writerow(['ID', 'Phish URL', 'Submitted', 'Valid?', 'Online?'])

    for page_num in range(0, 3000): 
        url = base_url.format(page_num)

        # Send a GET request to the URL
        response = requests.get(url)

        # Create a BeautifulSoup object to parse the HTML content of the response
        soup = BeautifulSoup(response.content, 'html.parser')

        # Find the table element containing the desired rows
        table = soup.find('table', class_='data')

        # Find all the rows in the table
        rows = table.find_all('tr')

        # Extract the values from each row and add them to a list of lists
        table_values = []
        for row in rows:
            row_values = [cell.text.strip() for cell in row.find_all('td', class_='value')]
            if row_values:
                table_values.append(row_values)

        # Write the table values to the CSV file, with each row in a separate row
        writer.writerows(table_values)

  
```
